```python
%pylab inline
import pylab as plt
file ='107647'
path_in = '/Users/dwaithe/Documents/collaborators/WaitheD/micro_vision/Faster-RCNN-TensorFlow-Python3.5/data/hek_peroxisome_all_class/2018/'
f = open(path_in+'labels/'+file+'.txt',"r")
img = plt.imread(path_in+"JPEGImages/"+file+".jpg")

width = img.shape[0]
height = img.shape[1]

lines =[]
while 1:
    line = f.readline()
    print(line)
    if not line:
        break
    lines.append(line)
figure()
imshow(img*5) 
for line in lines:
    ele = line.split(" ")
    xmin = float(ele[1])*width
    ymin = float(ele[2])*height
    wid = float(ele[3])*width
    hei = float(ele[4].split("\r")[0])*height
    #print(xmin,ymin,wid,hei)
    
    plta = [xmin-(wid//2),xmin-(wid//2),xmin+(wid//2),xmin+(wid//2),xmin-(wid//2)]
    pltb = [ymin-(hei//2),ymin+(hei//2),ymin+(hei//2),ymin-(hei//2),ymin-(hei//2)]
    
    plot(plta,pltb,'y-')
    
 
f.close()
```
