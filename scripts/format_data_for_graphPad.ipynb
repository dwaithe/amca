```python
##


folder = "/Users/dwaithe/Documents/collaborators/WaitheD/micro_vision/results/"

#Faster RCNN.
#conds = ["default/glycophorinA", "default/nucleopore","default/c127","default/neuroblastoma","default/eukaryote","default/peroxisome"]
#fname = "default_indlog_noflip.txt"
#fname = "default_indlog_flip.txt"
#fname = "default_glolog_noflip.txt"
#fname = "default_glolog_flip.txt"
#fname = "default_glo_withcollog.txt"
#fname = "default_ind_withcollog.txt"
fname = "darknet_glo_yolov3_nofliplog.txt"

#Darknet.
conds = ["cell - celltype", "nucleopore","c127","cell - neuroblastoma","eukaryote","peroxisome","cell - erythroblast"]

#fname = "darknet_indfliplog.txt"
#fname = "darknet_indnofliplog.txt"
#fname = "darknet_glolog.txt"
#fname = "darknet_glo_nofliplog.txt"
#fname = "darknet_glo_inc_colourlog.txt"
#fname = "darknet_indflip_colourlog.txt"

for cond in conds:
    with open(folder+fname) as f:
        content = f.readlines()

    # you may also want to remove whitespace characters like `\n` at the end of each line
    content = [x.strip() for x in content] 
    list_out =[]
    for line in content:
        sects = line.split("\t")
        if cond  in sects[1]:
            list_out.append(sects[5])
    num = list_out.__len__()//3
    count = 1
    print (cond)
    for ib in range(list_out.__len__()//3):
        print (list_out[ib],",",list_out[ib+num],",",list_out[ib+num*2])
        count +=1
```
